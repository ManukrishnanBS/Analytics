```python
%%sql
With track_qty As
    (select il.quantity,
        i.invoice_id,
        il.track_id
    FROM invoice_line as il
    INNER JOIN invoice i on il.invoice_id = i.invoice_id
    WHERE i.billing_country = "USA"),
qty_per_track As
    (Select tq.track_id,
            t.genre_id,
            sum(tq.quantity) tracks_sold
    from track_qty as tq
    inner join track t on t.track_id = tq.track_id
    group by tq.track_id)
    
Select * from qty_per_track;
```

```python
%%sql
With track_qty As
    (select il.quantity,
        i.invoice_id,
        il.track_id
    FROM invoice_line as il
    INNER JOIN invoice i on il.invoice_id = i.invoice_id
    WHERE i.billing_country = "USA"),
    
qty_per_track As
    (Select tq.track_id,
            t.genre_id,
            sum(tq.quantity) tracks_sold
    from track_qty as tq
    inner join track t on t.track_id = tq.track_id
    group by tq.track_id)

Select
g.name as genre_name,
    Sum(qt.tracks_sold) tracks_sold,
    Cast(tracks_sold as float)/(select sum(quantity)
        from invoice_line) as percentage
from genre as g
inner join qty_per_track qt on qt.genre_id = g.genre_id
group by g.name
order by tracks_sold desc;
    
```

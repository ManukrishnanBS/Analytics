```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
```

```python
recent_grads.plot('Sample_size', 'Unemployment_rate', kind='scatter')
```

```python
recent_grads.plot('Women', 'Median', kind='scatter')
```

```python
cols = ['Sample_size','Median','Employed','Full_time','ShareWomen','Unemployment_rate','Men','Women']
fig = plt.figure(figsize=(8,18))
for r in range(4):
    ax = fig.add_subplot(4,1,r+1)
    ax = recent_grads[cols[r]].hist(bins=20)
    ax.set_title(cols[r])
```

```python
cols = ['Sample_size','Median','Employed','Full_time','ShareWomen','Unemployment_rate','Men','Women']
fig = plt.figure(figsize=(8,18))
for r in range(4,8):
    ax = fig.add_subplot(4,1,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=30)
    ax.set_title(cols[r])
```
